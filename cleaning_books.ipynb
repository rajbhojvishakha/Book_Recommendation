```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem_text(text):
    words = text.split()

    stemmed_words = []

    for word in words:
        stemmed_words.append(ps.stem(word))

    return " ".join(stemmed_words)
```

```python
df['tags'] = df['tags'].str.lower()
```

```python
df['tags']=df['tags'].apply(stem_text)
```
